```python
# Importing all the necessary libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Calculates the cost of cultivation per acre based on fertilizer, pesticide, and water usage
data['Cost_of_Cultivation (USD/acre)'] = ((data['Fertilizer_Used(tons)'] * 500) + 
                                          (data['Pesticide_Used(kg)'] * 50) + 
                                          (data['Water_Usage(cubic meters)'] * 0.05)) / data['Farm_Area(acres)']
# Returns the first seven rows of the DataFrame
data.head(7)
```

```python
# Display unique crop types and their count
print(f"Unique Crop Types: {data['Crop_Type'].unique()}")
print(f"Number of Unique Crop Types: {data['Crop_Type'].nunique()}")
```

```python
# Shows summary statistics (like mean, min, max) for each column
data.describe().T
```

```python
cat_cols=data.select_dtypes(include=['object']).columns
num_cols = data.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
for col in num_cols:
    print(col)
    print('Skew :', round(data[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[col])
    plt.show()
```

```python

fig, axes = plt.subplots(3, 2, figsize=(18, 18))  # Create a 3x2 grid of subplots
fig.suptitle('Bar plot for all categorical variables in the farming dataset', fontsize=16)

# Plot bar plots for each categorical variable
sns.countplot(ax=axes[0, 0], x='Crop_Type', data=data, color='blue', 
              order=data['Crop_Type'].value_counts().index)
axes[0, 0].set_title('Crop Type')

sns.countplot(ax=axes[0, 1], x='Irrigation_Type', data=data, color='blue', 
              order=data['Irrigation_Type'].value_counts().index)
axes[0, 1].set_title('Irrigation Type')

sns.countplot(ax=axes[1, 0], x='Soil_Type', data=data, color='blue', 
              order=data['Soil_Type'].value_counts().index)
axes[1, 0].set_title('Soil Type')

sns.countplot(ax=axes[1, 1], x='Season', data=data, color='blue', 
              order=data['Season'].value_counts().index)
axes[1, 1].set_title('Season')

axes[2, 0].axis('off')
axes[2, 1].axis('off')

# Set tick rotation for better readability
axes[0, 0].tick_params(labelrotation=45)
axes[0, 1].tick_params(labelrotation=45)
axes[1, 0].tick_params(labelrotation=45)
axes[1, 1].tick_params(labelrotation=45)

# Adjust layout and display the plot
plt.tight_layout(rect=[0, 0, 1, 0.96])  
plt.show()
```

```python
# Sets the figure size for the plot to 13x17 inches
plt.figure(figsize=(13, 17))

# Creates pair plots to visualize relationships between numerical variables
# Drops 'Farm_ID' as it is just an identifier and not useful for analysis
sns.pairplot(data=data.drop(['Farm_ID'], axis=1))

# Displays the plot
plt.show()
```

```python
# Creates a figure and axis for plotting
fig, ax = plt.subplots()

# Plots a scatter plot to visualize the relationship between Yield and Water Usage
ax.scatter(data["Yield(tons)"], data["Water_Usage(cubic meters)"])

# Sets the title of the plot
ax.set_title("Scatter Plot of Yield vs Water Usage")

# Labels the x-axis as "Yield (tons)"
ax.set_xlabel("Yield (tons)")

# Labels the y-axis as "Water Usage (cubic meters)"
ax.set_ylabel("Water Usage (cubic meters)")

# Displays the figure
fig.show()
```

```python
# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Divide the 'Farm_Area(acres)' column into 7 bins (ranges)
area_bins = pd.cut(data["Farm_Area(acres)"], bins=7)

# Count the number of farms in each bin and sort by range
area_counts = area_bins.value_counts().sort_index()

# Create a bar plot to show the distribution of farm areas
area_counts.plot.bar(
    ax=ax,  # Use the created axis for plotting
    title="Farm Area Distribution",  # Set the title of the plot
    xlabel="Area Range (acres)",  # Label the x-axis
    ylabel="Count",  # Label the y-axis
)

# Show the plot
fig.show()
```
